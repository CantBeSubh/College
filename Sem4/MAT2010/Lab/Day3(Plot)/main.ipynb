```python
#plot a curve
import matplotlib.pyplot as plt
import numpy as np
pi=3.14*2
x = np.linspace(-pi,pi, 200)
y = np.sin(x)
z = np.exp(x)
fig, ax = plt.subplots()
ax.plot(x, y,'g--', linewidth=2.0)
ax.plot(x, z,'r', linewidth=1.0)
ax.set(xlim=(-pi, pi),   ylim=(-1, 1))
plt.show()
```

```python
x = np.linspace(0, 20, 100)
y = 2*np.sin(x)
y1= 2*np.cos(x)
y2= np.cos(10*x)
y3= np.sin(10*x)
fig, axs = plt.subplots(4,1)
pc = axs[0].plot(x, y, linewidth=2.0, color='r')
axs[0].set_title('2sin(x)')
co = axs[1].plot(x,y1, linewidth=2.0, color='g')
axs[1].set_title('2cos(x)')
pc1 = axs[2].plot(x, y2, linewidth=2.0, color='r')
axs[2].set_title('cos(2x)')
co = axs[3].plot(x,y3, linewidth=2.0, color='g')
axs[3].set_title('sin(2x)')
plt.show()
```

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T # transpose
z = x**2 + 3*y
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z,cmap='magma',edgecolor='none')
ax.set_title('Surface plot z=sqrt(x^2+y^2)')
plt.show()
```

```python
fig = plt.figure() 
ax = fig.add_subplot(1, 2, 1, projection='3d')
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T # transpose
z = (x*y**6 + y)
ax.plot_surface(x, y, z,cmap='magma',edgecolor='none')
ax.set_title('Surface plot z= x(y^6) + y')
ax = fig.add_subplot(1, 2, 2, projection='3d')
z2 = np.sin(2*x)
ax.plot_surface(x, y, z2,cmap='Purple',edgecolor='none')
ax.set_title('Surface plot z2 = sin(2x)')
plt.show()
```
